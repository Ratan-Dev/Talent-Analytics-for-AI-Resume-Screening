```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

# Set up the Chrome driver
driver = webdriver.Chrome()

page_urls = ["https://www.indeed.com/jobs?q=data+scientist&l=&from=searchOnHP&vjk=24e52ae053258ba6",
             "https://www.indeed.com/jobs?q=data+scientist&start=10&vjk=46777cb2bef680c3",
             "https://www.indeed.com/jobs?q=data+scientist&start=20&vjk=dcb82a18c468438b",
             "https://www.indeed.com/jobs?q=data+scientist&start=30&vjk=e3986b7c7151c666",
             "https://www.indeed.com/jobs?q=information+technology&l=&vjk=9f58e89e03e0c63e",
             "https://www.indeed.com/jobs?q=information+technology&start=10&vjk=045c96fccb4d3817",
             "https://www.indeed.com/jobs?q=information+technology&start=20&vjk=68218c874231bb12",
             "https://www.indeed.com/jobs?q=information+technology&start=30&vjk=17879e82da092b9e"]
titles, comps, des = [], [], []
count = 1

# Step 1: Use webdriver to open URL and create BeautifulSoup object to parse document
for page_url in page_urls:
    driver.get(page_url)
    time.sleep(5)  # Add a delay to ensure the page is fully loaded
    page = driver.page_source
    # Create BeautifulSoup object
    soup = BeautifulSoup(page, 'html.parser')

    job_postings = soup.find_all("div", class_="job_seen_beacon")

    for job in job_postings:
        logger.info(f"Scraping Job {count}")
        title = job.find("h2", class_="jobTitle").text.strip()
        comp = job.find("div", class_="company_location").text.strip()
        link = job.find("a", class_="jcs-JobTitle")
        if link:
            link_url = "https://www.indeed.com" + link["href"]
            driver.get(link_url)
            time.sleep(2)

            # Parse the page source of the full description page
            full_desc_soup = BeautifulSoup(driver.page_source, "html.parser")

            # Find the full job description
            description_elem = full_desc_soup.find("div", class_="jobsearch-JobComponent-description")
            description = description_elem.text.strip() if description_elem else "Not found"
            desc_idx = description.find("description")
            desc_idx_after = desc_idx + len("description")
            description = description[desc_idx_after:]

            titles.append(title)
            comps.append(comp)
            des.append(description)
            logger.success(f"Finished Scraping Job {count}")
            count += 1
            
        
# Quit the driver
driver.quit()
```

```python
df = pd.DataFrame(zip(titles, comps, des), columns=["Title", "Company/Location", "Description"])
```

```python
df.to_csv("ScrapedJobs.csv")
```
